```python
import datetime
import os
import pandas as pd
import numpy as np
from pprint import pprint
import requests
# Scrappings tools for downloading
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import re
```

```python
qualifiying_weather = pd.read_csv('./input/weather_cleaned.csv')
```

```python
qualifiying_weather.tail()
```

```python
weather_dict = {
    'weather_warm': ['clear sky', 'clear', 'warm', 'sunny', 'hot', 'fine'],
    'weather_cold': ['mild'],
    'weather_dry': ['dry'],
    'weather_wet': ['rain', 'rain at time', 'wet', 'light rain', 'rainy', 'heavy rain', 'humid', 'rainy'],
    'weather_cloudy': ['cloudy', 'overcast', 'mostly cloudy', 'partly cloud', 'partly cloudy'],
}
```

```python
weather_df = pd.DataFrame(columns=weather_dict.keys())
```

```python
weather_df
```

```python
for col in weather_df:
    weather_df[col] = qualifiying_weather["clean_weather"].map(lambda x: 1 if any(i in weather_dict[col] for i in x.lower().split(',')) else 0)
```

```python
weather_df
```

```python
weather_info = pd.concat([qualifiying_weather, weather_df], axis=1)
```

```python
if not os.path.exists('./output'):
    os.mkdir('./output')
weather_info.to_csv('./output/qualifiying_weather.csv', index = False)
```
