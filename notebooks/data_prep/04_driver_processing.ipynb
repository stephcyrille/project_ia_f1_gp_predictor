```python
import pandas as pd
```

## 1. Chargement du Dataset

```python
df_drivers = pd.read_csv("../../data/inputs/drivers.csv")
```

```python
df_drivers.head()
```

## 2. Retirer les colonnes inutilisées

```python
df_drivers.drop(['driverRef', 'url'], axis=1, inplace=True)
df_drivers.head()
```

## 3. Ajouter une étique uniquement sur les pilotes encore actifs

```python
df_drivers['driver_is_active'] = 0
```

```python
# Liste des pilotes actifs actuels
list_drivers = [
    ("Max", "Verstappen"),
    ("Logan", "Sargeant"),
    ("Lando", "Norris"),
    ("Pierre", "Gasly"),
    ("Sergio", "Pérez"),
    ("Fernando", "Alonso"),
    ("Charles", "Leclerc"),
    ("Lance", "Stroll"),
    ("Kevin", "Magnussen"),
    ("Yuki", "Tsunoda"),
    ("Alexander", "Albon"),
    ("Esteban", "Ocon"),
    ("Lewis", "Hamilton"),
    ("Carlos", "Sainz"),
    ("George", "Russell"),
    ("Oscar", "Piastri"),
    ("Guanyu", "Zhou"),
    ("Nico", "Hülkenberg"),
    ("Valtteri", "Bottas")
]
```

```python
# Ajouter l'étique en fonction de la correspondance nom et prénom du pilote dans le Dataset
for (name, surname) in list_drivers:
    df_drivers.loc[(df_drivers['forename'] == name) & (df_drivers['surname'] == surname), 'driver_is_active'] = 1

df_drivers[df_drivers.driver_is_active == 1].head()
```

## 4. Exportons le Dataset

```python
df_drivers.to_csv('../../data/outputs/03_drivers_processed.csv', index=False)
```
