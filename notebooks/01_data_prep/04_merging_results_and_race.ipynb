```python
import pandas as pd
```

## 1. Chargement du Dataset

```python
df_races = pd.read_csv('../../data/inputs/races.csv')
df_results = pd.read_csv('../../data/outputs/01_results_cleaned.csv')
```

```python
print(df_races.shape)
df_races.head()
```

```python
print(df_results.shape)
df_results.head()
```

## 2. Renomons la colonne positionOrder par race_rank 

```python
df_results.rename({'positionOrder': 'race_rank'}, axis=1, inplace=True)
```

## 3. Associons les deux jeux de données

```python
df_race_results = df_results.merge(df_races[['raceId', 'year', 'round', 'circuitId', 'name']], how='inner', on='raceId')
```

```python
print(df_race_results.shape)
df_race_results.head()
```

## 4. Exporter le Dataset

```python
df_race_results.to_csv("../../data/outputs/03_races_results.csv", index=False)
```
