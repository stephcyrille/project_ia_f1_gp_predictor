```python
%matplotlib inline
import pandas as pd
```

## 1. Loading the Data set

```python
df_races = pd.read_csv('../data/inputs/races.csv')
df_results = pd.read_csv('../data/outputs/01_results_cleaned.csv')
```

```python
print(df_races.shape)
df_races.head()
```

```python
print(df_results.shape)
df_results.head()
```

## 2. Merging the two Data set

```python
df_race_results = df_results.merge(df_races[['raceId', 'year', 'round', 'circuitId', 'name']], how='inner', on='raceId')
```

```python
print(df_race_results.shape)
df_race_results = df_race_results[df_race_results.year >= 2008]
```

```python
print(df_race_results.shape)
df_race_results.head()
```

```python
df_race_results.to_csv("../data/outputs/02_races_results.csv", index=False)
```
